## Mencari Validitas ID Number Customer

Langkah selanjutnya adalah melakukan filtering terhadap data untuk memastikan kevalidan dari data yang digunakan. Adapun format ketentuannya adalah customer yang terdata harus memiliki ID Number yang valid.





### Filtering Data

Pada proses filtering data ini, perlu dilakukan dua hal, diantaranya:
1.   Memfilter ID Number Customer sesuai dengan kriteria
2.   Memfilter duplikasi ID Number Customer 



**1. Memfilter ID Number Customer sesuai Kriteria**


Kriteria ID Number (Phone Number) Customer yang valid :
*   Panjang karakter adalah 11 - 12 karakter
*   Terdiri dari angka saja dan tidak diperbolehkan ada karakter selain angka
*   Diawali dengan angka 45 pada 2 digit pertama

```python
# Menambahkan kolom bantuan 'valid_ID'
df_clean['valid_ID'] = df_clean['customerID'].astype(str).str.match(r'(45\d{9,10})') 
df_clean
```

Penambahan kolom baru 'valid_ID' bertujuan untuk memeriksa apakah data pada baris tertentu memiliki ID Number yang valid (True) sesuai dengan kriteria yang ditentukan. 

Selanjutnya akan dilakukan filtering data dimana data yang ditampilkan hanya baris data yang memiliki valid_ID bernilai True. Setelah itu kolom 'valid_ID' bisa dihilangkan karena tidak dibutuhkan setelah filtering data selesai.

```python
# Memfilter data dan menghapus kolom 'valid_ID'
df_clean = df_clean[df_clean['valid_ID'] == True].drop('valid_ID', axis=1)
df_clean
```

Dari data diatas maka diketahui bahwa data tersebut hanya memiliki **7006 baris** data dengan **ID Number yang valid**. 

**2. Memfilter Duplikasi ID Number Customer**

Tahap ini berguna untuk memastikan bahwa tidak ada ID Number Customer yang duplikat. Adapun biasanya duplikasi ID Number ini tipenya :
*   Duplikasi dikarenakan inserting melebihi satu kali dengan nilai yang sama tiap kolomnya 
*   Duplikasi dikarenakan inserting beda periode pengambilan data




```python
# Drop duplicate rows
df_clean.drop_duplicates()
```

```python
# Drop duplicate ID Customer sorted by Periode
df_clean = df_clean.sort_values('UpdatedAt', ascending=False).drop_duplicates(['customerID'])
df_clean
```

Setelah dilakukan filtering untuk pengecekan kevalidasian data yang digunakan, maka tersisa **6993 baris** data dengan **ID Number** yang **valid** dan **tidak ada duplikasi**.

## Mengatasi Missing Values

Selanjutnya akan dilakukan pengecekan missing values pada data dengan tools yang terdapat pada packages pandas. Pengecekan missing values ini bertujuan untuk mendeteksi apakah ada data yang tergolong churn atau tidak.

Untuk mengatasi missing values ini dilakukan 2 proses, yakni :
1. Mengatasi Missing Values dengan penghapusan baris
2. Mengatasi Missing Values dengan pengisian nilai tertentu

**1. Mengatasi Missing Values dengan penghapusan baris**

Akan dilakukan penghapusan baris dari data-data yang tidak terdeteksi apakah termasuk *churn* atau tidak, dalam hal ini data yang dibuang adalah data yang tidak termasuk ke dalam golongan apapun. Diasumsikan data modeller hanya bisa menerima data yang benar ada flag *churn* nya atau tidak.

```python
# Mengecek jumlah missing values pada kolom Churn
df_clean['Churn'].isnull().sum()
```

Setelah dilakukan pengecekan terhadap kolom Churn, terdapat 43 baris data dengan missing value sehingga perlu dilakukan drop/ penghapusan baris data. 

```python
# Menghapus data dengan missing values
df_clean.dropna(subset=['Churn'], inplace=True)
```

```python
# Menampilkan total baris data
df_clean.shape
```

Dari data diatas terlihat bahwa data yang telah dilakukan drop missing value nya berubah menjadi **6950 baris** data.
